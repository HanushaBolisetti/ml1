```python
y=df1['price']
y
```

```python
x=df1.drop('price',axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_test.head(5)
```

```python
x_train.head()
```

```python
#to make a model we need to split the data into testing and training
#we make the model from processed data only
```

```python
y_train.shape
#y train-only price
```

```python
x_train.shape
#x train-all columns except price
```

```python
#models:
#linear regression,logistic regression,lasso,ridge,random forest,elastic
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()  
reg.fit(x_train,y_train) 
```

```python
ypred=reg.predict(x_test)
```

```python
ypred#predicted prices
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,ypred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(ypred,y_test)
```

```python
results=p.DataFrame(columns=['price','predicted'])
results['price']=y_test
results['predicted']=ypred
results=results.reset_index()
results['ID']=results.index
results.head(15)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x='ID',y='price',data=results.head(50))
sns.lineplot(x='ID',y='predicted',data=results.head(50))
```
