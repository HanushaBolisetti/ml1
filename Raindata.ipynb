```python
null_state=['ANDAMAN & NICOBAR ISLANDS','ARUNACHAL PRADESH','JAMMU & KASHMIR','WEST MADHYA PRADESH','COASTAL KARNATAKA','LAKSHADWEEP']
cols=['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']
for i in null_state:
    z=data1[data1['SUBDIVISION']== i]
    for j in cols:
        col_mean= z[j].mean()
        z[j].fillna(col_mean,inplace=True)
    data1.update(z)
print(data1.shape)
print(data1.isna().sum())
data1.head()
```

```python
A=data1[(data1['SUBDIVISION']=='ARUNACHAL PRADESH') | (data1['SUBDIVISION']=='KERALA')| (data1['SUBDIVISION']=='WEST RAJASTHAN')|(data1['SUBDIVISION']=='EAST RAJASTHAN')]
A
```

```python
A['JJAS']=A.apply(lambda row:row.JUN + row.JUL + row.AUG + row.SEP,axis=1)#JJAS=jun,jul,aug,sep
A
```

```python
A['OND']=A.apply(lambda row:row.OCT + row.NOV + row.DEC ,axis=1)#OND=oct,nov,dec
A
```

```python
A['ANNUAL_RAIN_fall']=A.apply(lambda row: row.JAN + row.FEB + row.MAR + row.APR + row.MAY + row.JUN + row.JUL + row.AUG + row.SEP + row.OCT + row.NOV + row.DEC,axis=1)
A
```

```python
df=A.filter(['SUBDIVISION','JJAS','OND','ANNUAL_RAIN_fall','YEAR'],axis=1)
df
```
